```python
import matplotlib.pyplot as plt
from PIL import Image

# 画像パスのリスト
image_paths = [
    "LSTM.jpg",
    "cLSTM.jpg",
    "vec_LSTM.jpg",
    "vec_cLSTM.jpg"
]

# 画像の読み込み
images = [Image.open(path) for path in image_paths]

# サブプロットの作成（縦4つ）
fig, axes = plt.subplots(4, 1, figsize=(6, 15))  # 横幅を狭く・高さを広く

# 画像を表示
for ax, img in zip(axes, images):
    ax.imshow(img)
    ax.axis('off')  # 軸非表示

# 縦の間隔（hspace）を調整：0.05くらいが自然
plt.subplots_adjust(hspace=0.02)

#plt.tight_layout()
plt.savefig('preds_plot.png', dpi=600, format='png')
plt.show()
```
